### Ley de potencia

Se emplea el método estadístico más robusto para detectar leyes de potencia:

**Clauset–Shalizi–Newman (2009): "Power-Law Distributions in Empirical Data".**

Resultados:

- **k_min óptimo:** 250  
- **Exponente α:** 2.54  
- **Tamaño de la cola:** 9141 nodos  
- **KS mínimo:** 0.0048  

Interpretación:

- El grafo sigue una ley de potencia genuina en su cola.
- Exponentes entre 2 y 3 son característicos de redes “scale-free”.
- Esto explica la presencia de superhubs, sensibilidad a ataques y robustez a fallas aleatorias.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from scipy.stats import linregress

# Obtener grados
degrees = np.array([d for _, d in G.degree()])
degrees = degrees[degrees > 0]  # evitar log(0)

# Contar frecuencia por grado (PMF)
degree_counts = Counter(degrees)
deg = np.array(list(degree_counts.keys()))
freq = np.array(list(degree_counts.values()), dtype=float)
pmf = freq / freq.sum()

# Usar solo la cola (grados grandes)
k_min = 20
mask = deg >= k_min
deg_tail = deg[mask]
pmf_tail = pmf[mask]

# Log transform
log_k = np.log10(deg_tail)
log_p = np.log10(pmf_tail)

# Ajuste lineal
slope, intercept, r_value, _, _ = linregress(log_k, log_p)
gamma = -slope

print(f"Exponente gamma estimado: {gamma:.4f}")
print(f"R^2 del ajuste: {r_value**2:.4f}")

# Límite inferior razonable para evitar puntos extremadamente ruidosos
pmf_threshold = 1e-5  # puedes subirlo a 1e-4 si quieres la cola más corta
mask_valid = pmf_tail >= pmf_threshold

deg_fit = deg_tail[mask_valid]  # grados donde sí hay datos
x_fit = np.log10(deg_fit)
y_fit = intercept + slope * x_fit

# ---- Graficar ----
plt.figure(figsize=(10,6))
plt.scatter(deg_tail, pmf_tail, s=8, alpha=0.4, label="Datos (cola)")

# Línea roja recortada
plt.plot(deg_fit, 10**y_fit,
         color="red", linewidth=2, label=f"Ajuste lineal (γ={gamma:.2f})")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Grado (log)")
plt.ylabel("PMF (log)")
plt.title("Estimación del exponente γ de la ley de potencia (recta recortada)")
plt.legend()
plt.show()
```

**Método: Clauset–Shalizi–Newman (2009): "Power-Law Distributions in Empirical Data"**

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Obtener grados
degrees = np.array([d for _, d in G.degree()])
degrees = degrees[degrees > 0]   # quitamos ceros por seguridad

print("N nodos:", len(degrees))
print("grado min, max:", degrees.min(), degrees.max())
```

```python
# 2. Funciones auxiliares

def mle_alpha_discrete(data, k_min):
    """
    Estimador MLE para la ley de potencia discreta
    usando la aproximación de Clauset (x_min - 0.5).
    data: array de grados (solo cola, >= k_min)
    """
    data = np.array(data, dtype=float)
    n = len(data)
    return 1.0 + n / np.sum(np.log(data / (k_min - 0.5)))


def ks_distance_discrete(data, k_min, alpha):
    """
    Distancia KS entre CDF empírica y CDF teórica
    para una ley de potencia discreta.
    """
    data = np.array(data, dtype=int)
    data = data[data >= k_min]
    n = len(data)

    # Soporte de la distribución: todos los enteros desde k_min hasta max(data)
    k_vals = np.arange(k_min, data.max() + 1)

    # PMF ~ k^(-alpha)
    pmf = k_vals ** (-alpha)
    pmf = pmf / pmf.sum()

    # CDF teórica F(k) = P(X <= k)
    cdf_theo = np.cumsum(pmf)

    # CDF empírica en los puntos únicos ordenados
    data_sorted = np.sort(data)
    unique_data, counts = np.unique(data_sorted, return_counts=True)
    cdf_emp = np.cumsum(counts) / float(n)

    # Para comparar, necesitamos F_theo en los mismos puntos unique_data
    # Mapeamos cada valor x de unique_data al índice correspondiente en k_vals
    idx = unique_data - k_min
    cdf_theo_at_data = cdf_theo[idx]

    # Distancia KS: sup |F_emp - F_teo|
    ks = np.max(np.abs(cdf_emp - cdf_theo_at_data))
    return ks
```

```python
# 3. Búsqueda de k_min y alpha

def fit_power_law_clauset(degrees,
                          kmin_min=20,
                          kmin_max=500,
                          min_tail_size=100):
    """
    Implementación sencilla del método de Clauset-Shalizi-Newman para
    ajustar ley de potencia discreta.

    - degrees: array con todos los grados (>0)
    - kmin_min, kmin_max: rango de búsqueda de k_min
    - min_tail_size: tamaño mínimo de la cola para considerar k_min
    """
    degrees = np.array(degrees, dtype=int)
    unique_deg = np.unique(degrees)

    # candidatos de k_min dentro del rango y presentes realmente
    candidates = unique_deg[(unique_deg >= kmin_min) & (unique_deg <= kmin_max)]
    print("Candidatos k_min:", len(candidates))

    best_ks = np.inf
    best_kmin = None
    best_alpha = None
    best_tail = None

    for k_min in candidates:
        tail = degrees[degrees >= k_min]
        n_tail = len(tail)
        if n_tail < min_tail_size:
            continue  # no hay suficiente muestra en la cola

        alpha = mle_alpha_discrete(tail, k_min)
        ks = ks_distance_discrete(tail, k_min, alpha)

        if ks < best_ks:
            best_ks = ks
            best_kmin = k_min
            best_alpha = alpha
            best_tail = tail

    return best_kmin, best_alpha, best_ks, best_tail


best_kmin, best_alpha, best_ks, tail = fit_power_law_clauset(degrees,
                                                             kmin_min=20,
                                                             kmin_max=500,
                                                             min_tail_size=200)

print(f"Mejor k_min: {best_kmin}")
print(f"Exponente gamma (alpha) MLE: {best_alpha:.4f}")
print(f"KS mínimo: {best_ks:.4f}")
print(f"Tamaño de la cola: {len(tail)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de la cola
tail = np.array(tail, dtype=int)
n_tail = len(tail)

# CCDF empírica
x_emp = np.sort(tail)
cdf_emp = np.arange(1, n_tail + 1) / float(n_tail)
ccdf_emp = 1.0 - cdf_emp + 1.0/n_tail

# CCDF teórica discreta
k_vals = np.arange(best_kmin, tail.max() + 1)
pmf = k_vals ** (-best_alpha)
pmf = pmf / pmf.sum()
cdf_theo = np.cumsum(pmf)
ccdf_theo = 1.0 - cdf_theo + pmf  # CCDF(k)

# Interpolación teórica en x_emp
idx = x_emp - best_kmin
ccdf_theo_at_emp = ccdf_theo[idx]

# RECORTE de la curva teórica para no extenderla demasiado
valid = ccdf_theo_at_emp > 1e-7   
x_fit = x_emp[valid]
y_fit = ccdf_theo_at_emp[valid]

# Gráfica
plt.figure(figsize=(10,6))

# CCDF empírica (puntos MORADOS)
plt.scatter(x_emp, ccdf_emp, s=14, alpha=0.6, 
            color="#141e76", edgecolors="none",
            label="CCDF empírica")

# CCDF teórica ajustada (línea roja recortada)
plt.plot(x_fit, y_fit, "r-", linewidth=2,
         label=f"Ley de potencia ajustada\n(alpha={best_alpha:.2f}, k_min={best_kmin})")

plt.xscale("log")
plt.yscale("log")

plt.xlabel("Grado k (log)", fontsize=12)
plt.ylabel("P(X ≥ k)  (log)", fontsize=12)
plt.title("Ajuste de ley de potencia (método Clauset–Shalizi–Newman)", fontsize=14)

plt.grid(True, which="both", ls="--", alpha=0.35)
plt.legend()
plt.tight_layout()
plt.show()
```

### Detección de comunidades 

La estructura comunitaria es un aspecto clave en redes complejas, ya que revela grupos de nodos más densamente conectados entre sí que con el resto del grafo.
Para identificar estas comunidades usamos el algoritmo Louvain, uno de los métodos más eficientes para grafos grandes.

Método empleado:

- A partir del grafo completo, aplicamos community_louvain.best_partition(G) para obtener un label de comunidad por nodo.

- Se almacena el identificador de comunidad en un diccionario {nodo: comunidad}.

- Finalmente agregamos esta información al DataFrame de nodos.

```python
# pip install python-louvain
```

```python
import community as community_louvain
import networkx as nx
import numpy as np

# Partición de Louvain (comunidad por cada nodo)
partition = community_louvain.best_partition(G)

# Añadir atributo 'community'
nx.set_node_attributes(G, partition, "community")
```

```python
from collections import defaultdict

communities = defaultdict(list)
for node, comm in partition.items():
    communities[comm].append(node)

print("Comunidades detectadas:", len(communities))
```

**Muestreo:** Debido al tamaño del grafo original (~168k nodos, ~6.8M edges), una visualización directa no es posible.
Por eso se aplicó un muestreo estructural, que conserva la forma general del grafo.

Estrategia:

- Elegir un subconjunto principal mediante ALEATORIO ESTRATIFICADO por comunidad.

- Agregar explícitamente los superhubs (nodos con mayor grado), ya que estos definen la estructura global.

- Construir el subgrafo inducido con NetworkX.

```python
import random

random.seed(42)

# nº de comunidades a usar
num_comms = 25

# escoger comunidades más grandes
largest_comms = sorted(communities.keys(), key=lambda c: len(communities[c]), reverse=True)[:num_comms]

sampled_nodes = []

for comm in largest_comms:
    nodes = communities[comm]
    
    # muestreo interno de cada comunidad
    k = min(120, len(nodes))
    sampled_nodes.extend(random.sample(nodes, k))

# Añadir hubs globales (PageRank top 50)
pagerank = nx.pagerank(G, alpha=0.85)
top_hubs = [n for n, _ in sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:50]]

sampled_nodes.extend(top_hubs)
sampled_nodes = list(set(sampled_nodes))  # sin duplicados

H = G.subgraph(sampled_nodes).copy()

print("Nodos del subgrafo:", H.number_of_nodes())
print("Edges del subgrafo:", H.number_of_edges())
```

**Gráfica 1:**

Esta visualización muestra el subgrafo obtenido mediante muestreo estratificado y con superhubs añadidos.
Los colores representan las comunidades obtenidas por Louvain.

Método:

- Layout usado: spring layout (fuerzas de atracción/repulsión).

- Color por comunidad (23 detectadas).

- Tamaño de nodos ligeramente mayor en hubs.

Interpretación:

- La red muestra una forma globular, indicando alta interconectividad.

- Los colores representan módulos del grafo: grupos de streamers más conectados entre sí.

- Los hubs (nodos grandes) aparecen en el núcleo, lo esperado en redes “scale-free”.

- La presencia de brazos y cúmulos periféricos sugiere clusters temáticos o lingüísticos

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 14))

# posiciones del layout
pos = nx.spring_layout(
    H,
    k=0.12,         # separación
    iterations=60,
    seed=42
)

# colores por comunidad
node_comms = [G.nodes[n]["community"] for n in H.nodes()]
unique_comms = list(set(node_comms))
color_map = {c:i for i,c in enumerate(unique_comms)}
mapped_colors = [color_map[c] for c in node_comms]

# tamaños de nodo (hubs más grandes)
node_sizes = [np.log1p(H.degree(n))*40 for n in H.nodes()]

# dibujar nodos
nx.draw_networkx_nodes(
    H, pos,
    node_size=node_sizes,
    node_color=mapped_colors,
    cmap="tab20",
    alpha=0.90
)

# dibujar edges
nx.draw_networkx_edges(
    H, pos,
    alpha=0.04,
    width=0.4,
    edge_color="gray"
)

plt.title("Subgrafo representativo de Twitch (muestral + comunidades + hubs)", fontsize=17)
plt.axis("off")
plt.show()
```

**Gráfica 2:**

Esta visualización proyecta los embeddings estructurales de los nodos (generados con Node2Vec) a 2D usando UMAP, donde cada punto es un streamer y el color marca su comunidad Louvain.

Método:

- Generar embeddings 64D con Node2Vec (paseos aleatorios sesgados).

- Reducir a 2D mediante UMAP (Uniform Manifold Approximation and Projection).

- Colorear según comunidad.

Interpretación:

- Los clusters corresponden a comunidades estructurales detectadas en la red.

- Dos núcleos centrales grandes (muchos colores mezclados) indican:

- Comunidades muy cercanas estructuralmente. O Louvain detecta muchas subcomunidades pequeñas dentro de una región densa.

Grupos periféricos bien separados suelen ser:

- Comunidades idiomáticas (RU, TR, ES, ZH…)

- Nichos temáticos

- Streamers muy conectados entre sí pero aislados del resto.

```python
# pip install node2vec umap-learn
```

```python
from node2vec import Node2Vec
import umap
import numpy as np
import matplotlib.pyplot as plt

# Entrenar Node2Vec
node2vec = Node2Vec(
    H,
    dimensions=64,
    walk_length=20,
    num_walks=150,
    workers=1
)
model = node2vec.fit(window=10, min_count=1, batch_words=4)

# Embeddings
nodes = list(H.nodes())
X = np.array([model.wv[str(n)] for n in nodes])

# UMAP 2D
reducer = umap.UMAP(n_neighbors=20, min_dist=0.1)
X_2d = reducer.fit_transform(X)

# Comunidades → colores
node_comms = np.array([G.nodes[n]["community"] for n in nodes])
unique = sorted(set(node_comms))
colormap = {c:i for i,c in enumerate(unique)}
colors = np.array([colormap[c] for c in node_comms])

plt.figure(figsize=(15,15))
plt.scatter(X_2d[:,0], X_2d[:,1], c=colors, cmap="tab20", s=12, alpha=0.85)
plt.title("Embeddings Node2Vec proyectados con UMAP (coloreado por comunidad)", fontsize=20)
plt.axis("off")
plt.show()
```

¿Por qué salen tantos colores mezclados en el centro?

Porque Louvain encontró 23 comunidades, muchas de ellas microcomunidades muy similares estructuralmente. UMAP agrupa según proximidad en embeddings, no según comunidad, entonces varias comunidades pequeñas terminan superpuestas.

De la siguiente tabla cada fila corresponde a una comunidad detectada dentro de la red, e incluye:

- size: número de nodos en la comunidad

- top_languages: los idiomas predominantes

- affiliate_rate: proporción de usuarios afiliados dentro de la comunidad

- mature_rate: proporción de contenido para adultos

- avg_views: vistas promedio (variable altamente sesgada)

- median_views: vistas medianas (más estable contra outliers)

- avg_life_days: actividad promedio en días

```python
import numpy as np
import statistics

summary = []

for c in comm_data:
    langs = pd.Series(comm_data[c]["language"]).value_counts().head(3)
    aff_rate = np.mean(comm_data[c]["affiliate"])
    mat_rate = np.mean(comm_data[c]["mature"])
    avg_views = np.mean(comm_data[c]["views"])
    median_views = np.median(comm_data[c]["views"])
    avg_life = np.mean(comm_data[c]["life_time"])

    summary.append({
        "community": c,
        "size": len(comm_data[c]["views"]),
        "top_languages": langs.to_dict(),
        "affiliate_rate": aff_rate,
        "mature_rate": mat_rate,
        "avg_views": avg_views,
        "median_views": median_views,
        "avg_life_days": avg_life
    })

pd.DataFrame(summary).sort_values("size", ascending=False)
```

### **Centralidades**

Se calcularon tres medidas fundamentales de centralidad para cada nodo:

- Degree Centrality

- PageRank

- Eigenvector Centrality

Estas métricas permiten identificar nodos relevantes (influencers) según diferentes criterios estructurales.

---

1. Degree Centrality

- Mide el número de conexiones directas.

- Los nodos con mayor grado tienden a ubicarse en el núcleo del grafo.

- Identifican streamers con gran exposición dentro de la plataforma.

2. PageRank

- Mide la importancia considerando el prestigio de los vecinos.

- A diferencia del grado, PageRank valora quién te sigue o conecta contigo.

- Los superhubs identificados poseen PageRank excepcionalmente alto.

- Es la mejor métrica para medir influencia global.

3. Eigenvector Centrality

- También mide prestigio, pero enfatiza conexiones entre nodos influyentes.

- Identifica líderes dentro de subcomunidades.

- Nodos con alto eigenvector suelen estar dentro de clusters temáticos.

### 1. Cálculo de centralidades

```python
import networkx as nx
import pandas as pd
import numpy as np

# Usaremos tu subgrafo H
Gsub = H

# 1. Degree centrality
deg_cent = nx.degree_centrality(Gsub)

# 2. Approx Betweenness centrality (k=100 nodos muestra)
bet_cent = nx.betweenness_centrality(Gsub, k=100, seed=42)

# 3. PageRank
pr_cent = nx.pagerank(Gsub, alpha=0.85)

# 4. Eigenvector centrality (método iterativo)
eig_cent = nx.eigenvector_centrality(Gsub, max_iter=300)

# Construir DataFrame
centrality_df = pd.DataFrame({
    "node": list(Gsub.nodes()),
    "degree_centrality": [deg_cent[n] for n in Gsub.nodes()],
    "betweenness_centrality": [bet_cent[n] for n in Gsub.nodes()],
    "pagerank": [pr_cent[n] for n in Gsub.nodes()],
    "eigenvector": [eig_cent[n] for n in Gsub.nodes()],
})
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import math

def eval_reg(name, model, X_tr, X_te, y_tr, y_te, scaled=False):
    if scaled:
        y_pred_tr = model.predict(X_tr)
        y_pred_te = model.predict(X_te)
    else:
        y_pred_tr = model.predict(X_tr)
        y_pred_te = model.predict(X_te)

    rmse_tr = math.sqrt(mean_squared_error(y_tr, y_pred_tr))
    rmse_te = math.sqrt(mean_squared_error(y_te, y_pred_te))
    mae_te  = mean_absolute_error(y_te, y_pred_te)
    r2_te   = r2_score(y_te, y_pred_te)

    print(f"=== {name} ===")
    print(f"Train RMSE (log): {rmse_tr:.4f}")
    print(f"Test  RMSE (log): {rmse_te:.4f}")
    print(f"Test  MAE  (log): {mae_te:.4f}")
    print(f"Test  R^2       : {r2_te:.4f}")
    print()

# LinearReg usa escalado
eval_reg("LinearRegression", linreg, X_train_scaled, X_test_scaled, y_train, y_test, scaled=True)

# RF y XGB usan X sin escalar
eval_reg("RandomForest", rf, X_train, X_test, y_train, y_test)
eval_reg("XGBoost", xgbr, X_train, X_test, y_train, y_test)
```

La detección de comunidades mostró una organización modular clara: clusters asociados principalmente a idioma, temática y patrones de audiencia. Las proyecciones UMAP de embeddings Node2Vec reforzaron esta estructura, revelando grupos cohesivos y un núcleo central multicolor donde varias comunidades se entremezclan, indicando un espacio de alta interacción.

El estudio de centralidades evidenció que las métricas basadas en influencia como PageRank y eigenvector describen mejor la importancia de un canal que el simple número de seguidores. De hecho, los canales más influyentes no siempre son los de mayor grado, sino aquellos conectados con otros usuarios relevantes.

Finalmente, el modelo de predicción de popularidad mostró que la posición estructural del canal dentro de la red (especialmente PageRank) es un factor determinante para explicar sus views. Aunque los datos son muy dispersos, la regresión logró capturar una parte significativa de su variabilidad, confirmando que la estructura de la red es un predictor clave de popularidad.

En conjunto, el proyecto demuestra cómo el uso integrado de análisis de redes, embeddings, detección de comunidades y modelos predictivos ofrece una comprensión profunda del ecosistema digital de Twitch y de los factores que moldean la visibilidad y relevancia de los canales.
